```python
def tampil_papan(papan):
    print(f" {papan[7]} | {papan[8]} | {papan[9]} ")
    print("-----------")
    print(f" {papan[4]} | {papan[5]} | {papan[6]} ")
    print("-----------")
    print(f" {papan[1]} | {papan[2]} | {papan[3]} ")

def masukan_pemain():
    tanda = ''
    while not (tanda == 'X' or tanda == 'O'):
        tanda = input("Pemain 1: Apakah Anda ingin menjadi X atau O? ").upper()

    if tanda == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')

def letakkan_tanda(papan, tanda, posisi):
    papan[posisi] = tanda

def cek_kemenangan(papan, tanda):
    return ((papan[7] == papan[8] == papan[9] == tanda) or
            (papan[4] == papan[5] == papan[6] == tanda) or
            (papan[1] == papan[2] == papan[3] == tanda) or
            (papan[7] == papan[4] == papan[1] == tanda) or
            (papan[8] == papan[5] == papan[2] == tanda) or
            (papan[9] == papan[6] == papan[3] == tanda) or
            (papan[7] == papan[5] == papan[3] == tanda) or
            (papan[9] == papan[5] == papan[1] == tanda))

def cek_kosong(papan, posisi):
    return papan[posisi] == ' '

def cek_papan_penuh(papan):
    return ' ' not in papan[1:]

def pilihan_pemain(papan):
    posisi = 0
    while posisi not in range(1, 10) or not cek_kosong(papan, posisi):
        posisi = int(input("Pilih posisi Anda selanjutnya: (1-9) "))

    return posisi

def main_ulang():
    return input("Apakah Anda ingin bermain lagi? Masukkan Ya atau Tidak: ").lower().startswith('y')

# Logika utama permainan
print("Selamat datang di Tic Tac Toe!")

while True:
    papan_permainan = [' '] * 10
    pemain1_tanda, pemain2_tanda = masukan_pemain()
    giliran = 'Pemain 1'

    main_game = input("Apakah Anda siap untuk bermain? Masukkan Ya atau Tidak: ")

    if main_game.lower()[0] == 'y':
        game_on = True
    else:
        game_on = False

    while game_on:
        if giliran == 'Pemain 1':
            tampil_papan(papan_permainan)
            posisi = pilihan_pemain(papan_permainan)
            letakkan_tanda(papan_permainan, pemain1_tanda, posisi)

            if cek_kemenangan(papan_permainan, pemain1_tanda):
                tampil_papan(papan_permainan)
                print("Selamat! Pemain 1 menang!")
                game_on = False
            else:
                if cek_papan_penuh(papan_permainan):
                    tampil_papan(papan_permainan)
                    print("Permainan seri!")
                    break
                else:
                    giliran = 'Pemain 2'

        else:
            tampil_papan(papan_permainan)
            posisi = pilihan_pemain(papan_permainan)
            letakkan_tanda(papan_permainan, pemain2_tanda, posisi)

            if cek_kemenangan(papan_permainan, pemain2_tanda):
                tampil_papan(papan_permainan)
                print("Selamat! Pemain 2 menang!")
                game_on = False
            else:
                if cek_papan_penuh(papan_permainan):
                    tampil_papan(papan_permainan)
                    print("Permainan seri!")
                    break
                else:
                    giliran = 'Pemain 1'

    if not main_ulang():
        break
```
